```python
#Analisando Casos Confirmados de covid-19
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
    
urlretrieve(url, 'datasets/global_cases_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_covid19.csv')
```

```python
df_covid
```

```python
#Excluindo os dados que não são relevantes (latitude e longitude)
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True) 
```

```python
df_covid
```

```python
#Agrupando o dataframe pelos paises
df_country = df_covid.groupby('Country/Region').sum()
```

```python
df_country
```

```python
#Acessando uma linha 
df_country.loc['Brazil']
```

```python
#Criando uma variavel para armazenar informações expecificas 
#nesse exemplo vou criar uma variavel para armazenar as datas referentes ao pais Brasil

date = df_country.loc['Brazil'].index
cases = df_country.loc['Brazil'].values
```

```python
#Conjunto de todas as datas referentes ao Brasil
date
```

```python
#Conjunto de todos os casos referentes ao Brasil
cases
```

```python
#Grafico de barra
plt.figure(figsize=(14,8))
plt.bar(date,cases)
```

```python
df_country.head()
```

```python
s_brazil = df_country.loc['Brazil']
```

```python
s_brazil
```

```python
# pegamos somente onde s_brazil for maior que 0 ,ou seja, excluindo os dias em que o brasil não tinha nenhum caso  
s_brasil = s_brazil[s_brazil > 0]
```

```python
s_brasil
```

```python
s_brasil.index
```

```python
s_brasil.values
```

```python
#Considerando a data onde ocorreu o primeiro caso no Brasil
tam = len(s_brasil)


plt.figure(figsize=(14,8)) #Editando a imagem 
plt.xticks(rotation=60) #melhorando a visualizaação 
plt.bar(s_brasil.index[tam-30:tam], s_brasil.values[tam-30:tam]) #coletanto os 3 últimos valores
plt.title('Total de casos confirmados de COVID-19 no Brasil') #titulo 
```
